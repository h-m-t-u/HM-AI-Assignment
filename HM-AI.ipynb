Prompt: make a plot of the used_pin_number column with the color set to the fraud column, specifically red and green

```python
# Define a custom color palette for fraud
fraud_palette = {0: 'green', 1: 'red'}

# Create a count plot of used_pin_number colored by fraud
plt.figure(figsize=(6, 4))
sns.countplot(
    data=transaction_data_cleaned,
    x='used_pin_number',
    hue='fraud',
    palette=fraud_palette
)

# Add title and labels
plt.title('Used PIN Number Distribution by Fraud Status', fontsize=14)
plt.xlabel('Used PIN Number (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Fraud', labels=['Non-Fraud', 'Fraud'])

# Display the plot
plt.show()
```

Prompt: make a plot that shows the count of the column used_pin_number with the ones that ones that are fraud are red and not are green. It shows the percentage of fraudulent transactions by usage of pin

```python
# Calculate counts and percentages
pin_fraud_counts = transaction_data_cleaned.groupby(['used_pin_number', 'fraud']).size().reset_index(name='count')

# Pivot the table for easier plotting
pin_fraud_pivot = pin_fraud_counts.pivot(index='used_pin_number', columns='fraud', values='count').fillna(0)

# Calculate percentages
pin_fraud_percent = pin_fraud_pivot.div(pin_fraud_pivot.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
pin_fraud_percent.plot( 
    kind='bar', 
    stacked=True, 
    color={0: 'green', 1: 'red'}, 
    figsize=(6, 4)
)

# Add labels and title
plt.title('Percentage of Fraud by Used PIN Number', fontsize=14)
plt.xlabel('Used PIN Number (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Percentage of Transactions', fontsize=12)
plt.legend(title='Fraud', labels=['Non-Fraud', 'Fraud'])

plt.show()
```

Prompt: make a plot that shows a percentage plot to show the percentage of fraudulent transactions when a chip was used or not that ones that are fraud are red and not are green. It shows the percentage of fraudulent transactions by usage of chip

```python
# Calculate counts of fraud vs non-fraud grouped by used_chip
chip_fraud_counts = transaction_data_cleaned.groupby(['used_chip', 'fraud']).size().reset_index(name='count')

# Pivot the table for easier plotting
chip_fraud_pivot = chip_fraud_counts.pivot(index='used_chip', columns='fraud', values='count').fillna(0)

# Calculate percentages
chip_fraud_percent = chip_fraud_pivot.div(chip_fraud_pivot.sum(axis=1), axis=0) * 100

# Plot stacked bar chart of percentages
chip_fraud_percent.plot(
    kind='bar', 
    stacked=True, 
    color={0: 'green', 1: 'red'}, 
    figsize=(6, 4)
)

# Add labels and title
plt.title('Percentage of Fraudulent Transactions by Chip Usage', fontsize=14)
plt.xlabel('Used Chip (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Percentage of Transactions', fontsize=12)
plt.legend(title='Fraud', labels=['Non-Fraud', 'Fraud'])

# Show the plot
plt.show()
```
